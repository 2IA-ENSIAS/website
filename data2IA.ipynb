```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
# select theme for seaborn
sns.set_theme(style="ticks")
# selec color palette
sns.set_palette("autumn")
```

```python
data = pd.read_csv('2IA_data.csv', sep=',')
```

```python
data.head()
```

```python
data.columns
```

```python
# drop the comuns Horodateur and Adresse e-mail
data.drop(['Horodateur', 'Adresse e-mail'], axis=1, inplace=True)
```

```python
data.columns
```

```python
# split data into 3 dataframes (one for each three questions)
data1A = data.iloc[:, 0:3]
data2A = data.iloc[:, 3:6]
data3A = data.iloc[:, 6:]
```

```python
# rename columns
data1A.columns = ['company', 'position', 'fields']
data2A.columns = ['company', 'position', 'fields']
data3A.columns = ['company', 'position', 'fields']
```

```python
# drop the rows with NaN values
data1A.dropna(inplace=True)
data2A.dropna(inplace=True)
data3A.dropna(inplace=True)
```

```python
# reset index
data1A.reset_index(drop=True, inplace=True)
data2A.reset_index(drop=True, inplace=True)
data3A.reset_index(drop=True, inplace=True)
```

```python
# make all the strings lowercase
for i in range(len(data1A)):
    data1A.loc[i, 'company'] = data1A.loc[i, 'company'].lower()
    data1A.loc[i, 'position'] = data1A.loc[i, 'position'].lower()
    data1A.loc[i, 'fields'] = data1A.loc[i, 'fields'].lower()

for i in range(len(data2A)):
    data2A.loc[i, 'company'] = data2A.loc[i, 'company'].lower()
    data2A.loc[i, 'position'] = data2A.loc[i, 'position'].lower()
    data2A.loc[i, 'fields'] = data2A.loc[i, 'fields'].lower()

for i in range(len(data3A)):
    data3A.loc[i, 'company'] = data3A.loc[i, 'company'].lower()
    data3A.loc[i, 'position'] = data3A.loc[i, 'position'].lower()
    data3A.loc[i, 'fields'] = data3A.loc[i, 'fields'].lower()
```

```python
# plot the distribution of positions in each dataframe using seaborn
sns.countplot(data1A['position'])
plt.title('Positions in 1A')
plt.xticks(rotation=90)
plt.show()

sns.countplot(data2A['position'])
plt.title('Positions in 2A')
plt.xticks(rotation=90)
plt.show()

sns.countplot(data3A['position'])
plt.title('Positions in 3A')
plt.xticks(rotation=90)
plt.show()
```

```python
# split the fields column into a list of fields
data1A['fields'] = data1A['fields'].str.split(', ')
data2A['fields'] = data2A['fields'].str.split(', ')
data3A['fields'] = data3A['fields'].str.split(', ')
```

```python
# function to count the number of occurences of each field in a dataframe
def count_fields(data):
    fields = {}
    for i in range(len(data)):
        for j in range(len(data.loc[i, 'fields'])):
            if data.loc[i, 'fields'][j] in fields:
                fields[data.loc[i, 'fields'][j]] += 1
            else:
                fields[data.loc[i, 'fields'][j]] = 1
    return fields


# count the number of occurences of each field in each dataframe
fields1A = count_fields(data1A)
fields2A = count_fields(data2A)
fields3A = count_fields(data3A)

# sort the fields by number of occurences
fields1A = dict(sorted(fields1A.items(), key=lambda item: item[1], reverse=True))
fields2A = dict(sorted(fields2A.items(), key=lambda item: item[1], reverse=True))
fields3A = dict(sorted(fields3A.items(), key=lambda item: item[1], reverse=True))

# plot the distribution of fields in each dataframe using seaborn and pie charts
plt.pie(fields1A.values(), labels=fields1A.keys(), autopct='%1.1f%%', startangle=90)
plt.title('Fields in 1A')
plt.show()

plt.pie(fields2A.values(), labels=fields2A.keys(), autopct='%1.1f%%', startangle=90)
plt.title('Fields in 2A')
plt.show()

plt.pie(fields3A.values(), labels=fields3A.keys(), autopct='%1.1f%%', startangle=90)
plt.title('Fields in 3A')
plt.show()
```

```python
# plot the occurences of each company in all three dataframes
# concatenate the three dataframes
figure = plt.figure(figsize=(10, 20))
data = pd.concat([data1A, data2A, data3A], ignore_index=True)
sns.countplot(data['company'], order=data['company'].value_counts().index)
plt.title('Companies')
plt.xticks(rotation=90)
plt.show()
```
